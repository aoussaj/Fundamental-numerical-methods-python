```python
import math

# Definition of function
def f(x):
    return math.cos(x) - x

# Enter parameters
xLeft = float(input('Enter xLeft: '))
xRight = float(input('Enter xRight: '))
Tolerance = float(input('Enter Tolerance: '))

# Initial values
i = 0  # iteration counter

# Loop with break condition
while True:
    xMiddle = (xRight + xLeft) / 2

    if f(xLeft) * f(xMiddle) < 0:
        xRight = xMiddle
    else:
        xLeft = xMiddle

    Error = abs((xRight - xLeft) / xMiddle)
    i += 1

    # stopping condition
    if Error <= Tolerance:
        break

# Prints the results
print(f'Root found at {xMiddle:.8f}')
print('Number of iterations:', i)
```
