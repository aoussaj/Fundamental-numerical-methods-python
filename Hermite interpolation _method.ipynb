```python
import sympy as sp
import numpy as np

x = sp.symbols("x")

choice = int(input("Choose function: 1 (exp), 2 (1/(1+x^2)): "))

# choose nodes depending on function
if choice == 1:
    F = [-1, 0.5, 1.5, 2]
    f = lambda t: sp.exp(t)
    f_prime = lambda t: sp.exp(t)

elif choice == 2:
    F = list(range(-5, 6))
    f = lambda t: 1 / (1 + t**2)
    f_prime = lambda t: -2*t / (1 + t**2)**2

else:
    print("Invalid choice")
    exit()

N = len(F)

P = 0  # symbolic polynomial

for i in range(N):
    # build L_i(x)
    L_i = 1
    for j in range(N):
        if i != j:
            L_i *= (x - F[j]) / (F[i] - F[j])

    # Hermite basis functions
    h_i = (1 - 2*(x - F[i]) * f_prime(F[i])) * (L_i**2)
    hh_i = (x - F[i]) * (L_i**2)

    # accumulate polynomial
    P += h_i * f(F[i]) + hh_i * f_prime(F[i])

# simplify result
P = sp.simplify(P)

print("\nHermite polynomial:")
print(P)
```
