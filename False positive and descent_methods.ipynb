```python
import math

# Function definition
def f(x):
    return math.cos(x) - x

def get_float(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Please enter a valid number.")

def false_position():
    xL = get_float('Enter xLeft: ')
    xR = get_float('Enter xRight: ')
    Tol = get_float('Enter Tolerance: ')

    i = 0
    Error = Tol + 1
    xn = xL - 1

    fL = f(xL)
    fR = f(xR)

    if fL * fR > 0:
        print("Invalid interval: f(xL) and f(xR) must have opposite signs.")
        return

    while Error > Tol:
        xn1 = (xL * fR - xR * fL) / (fR - fL)
        fNew = f(xn1)

        i += 1
        print(f"i= {i:3} xn1= {xn1:.8f} xL= {xL:.8f} xR= {xR:.8f}")

        if fL * fNew < 0:
            xR = xn1
            fR = fNew
        else:
            xL = xn1
            fL = fNew

        Error = abs((xn1 - xn) / xn1)
        xn = xn1

    print(f'Root found at {xn1:.8f}')
    print('Number of iterations:', i)

def secant():
    xim1 = get_float('Enter x0: ')
    xi = get_float('Enter x1: ')
    Tol = get_float('Enter Tolerance: ')

    i = 0
    Error = Tol + 1

    f_im1 = f(xim1)
    f_i = f(xi)

    while Error > Tol:
        if f_i == f_im1:
            print("Division by zero encountered.")
            return

        xn1 = (xim1 * f_i - xi * f_im1) / (f_i - f_im1)
        f_new = f(xn1)

        i += 1
        print(f"i= {i:3} xn1= {xn1:.8f}")

        Error = abs((xn1 - xi) / xn1)

        xim1, f_im1 = xi, f_i
        xi, f_i = xn1, f_new

    print(f'Root found at {xn1:.8f}')
    print('Number of iterations:', i)

# Main menu
while True:
    print("\nOptions: (1) False-position ; (2) Secant ; (3) Exit")
    
    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input.")
        continue

    if choice == 1:
        false_position()
    elif choice == 2:
        secant()
    elif choice == 3:
        break
    else:
        print("Invalid choice")
```
