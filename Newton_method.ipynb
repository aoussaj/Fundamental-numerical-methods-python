```python
import math

# Definition of functions
def f1(x):
    return math.cos(x) - x

def f1_prime(x):
    return -1 * math.sin(x) - 1

def f2(x):
    return (6435 * x**8 - 12012 * x**6 + 6930 * x**4 - 1260 * x**2 + 35) / 128

def f2_prime(x):
    return (8 * 6435 * x**7 - 6 * 12012 * x**5 + 4 * 6930 * x**3 - 2 * 1260 * x) / 128

def f3(x):
    return x**3 - 169

def f3_prime(x):
    return 3 * x**2

# Choose function
print("Choose the function:")
print("(1) cos(x) - x ; (2) P(x) ; (3) x^3 - 169")
choice = int(input("Enter your choice: "))

if choice == 1:
    f = f1
    f_prime = f1_prime
elif choice == 2:
    f = f2
    f_prime = f2_prime
elif choice == 3:
    f = f3
    f_prime = f3_prime
else:
    print("Invalid choice, cos(x) - x is taken by default")
    f = f1
    f_prime = f1_prime

# Enter parameters
x_n = float(input("Enter x0: "))
Tolerance = float(input("Enter Tolerance: "))

# Initial values
i = 0
Error = Tolerance + 1

# Newton-Raphson method
while Error > Tolerance:
    x_n1 = x_n - (f(x_n) / f_prime(x_n))
    Error = abs((x_n1 - x_n) / x_n1)
    i += 1
    print(f"i = {i}, x_n+1 = {x_n1:.8f}")
    x_n = x_n1

# Results
print(f"Root found at {x_n1:.8f}")
print("Number of iterations:", i)
```
